```python
class Compra():
    def __init__(self, vendedor, data, valor):
        self.vendedor = vendedor
        self.data = data
        self.valor = valor

class Pessoa():
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        
    def __str__(self):
        return f"Nome: {self.nome} - Idade: {self.idade}"
    
    def isAdult(self):
        if self.idade >= 18 and self.idade < 60:
            return True
        return False
    
class Vendedor(Pessoa):
    def __init__(self, nome, idade, salario):
        super().__init__(nome, idade)
        self.salario = salario
        
class Cliente(Pessoa):
    compras = []
    
    def __init__(self, nome, idade):
        super().__init__(nome, idade)
        self.compras = []
        
    def registrar_compra(self, compra):
        self.compras.append(compra)
        
    def get_data_ultima_compra(self):
        return self.compras[-1].data
    
    def total_compras(self):
        total = 0
        for compra in self.compras:
            print(compra.valor)
            total += compra.valor
            
        return total
    
if __name__ == "__main__":
    guilherme = Cliente("Guilherme", 25)
    claudio = Cliente("Cláudio", 37)
    
    augusto = Vendedor("Augusto", 62, 1200.00)
    
    compra_1_guilherme = guilherme.registrar_compra(Compra(augusto, "2023-8-11", 30.55))
    compra_2_guilherme = guilherme.registrar_compra(Compra(augusto, "2023-8-13", 70.00))
    compra_3_guilherme = guilherme.registrar_compra(Compra(augusto, "2023-8-13", 22.00))
    
    compra_1_claudio = claudio.registrar_compra(Compra(augusto, "2023-8-9", 12.70))
    
    print("Última compra do Guilherme: ", guilherme.get_data_ultima_compra())
    print("Guilherme é um adulto? ", guilherme.isAdult())
    print("Total gasto por Guilherme: ", guilherme.total_compras())
    
    print("Última compra do Cláudio: ", claudio.get_data_ultima_compra())
    print("Cláudio é um adulto? ", claudio.isAdult())
    print("Total gasto por Cláudio: ", claudio.total_compras())
    
    print("Augusto é um adulto? ", augusto.isAdult())
    
    print(guilherme)
    print(claudio)
    print(augusto)
```
